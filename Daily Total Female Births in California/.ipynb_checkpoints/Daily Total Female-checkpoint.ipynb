```python
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.tsa.stattools import adfuller
```

```python
data = pd.Series.from_csv('daily-total-female-births-in-ca.csv', header=0)
data.head(10)
```

```python
data.plot(figsize=(10,7))
plt.grid(True)
plt.show()
```

```python
rol_mean = data.rolling(window=12).mean()
rol_std = data.rolling(window=12).std()
print(rol_mean, rol_std)
```

```python
original = plt.plot(data, color='blue', label='Original')
mean = plt.plot(data, color='red', label='Rolling Mean')
std = plt.plot(data, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.grid(True)
plt.show(block=False)
```

```python
print('Results of Dickey-Fuller Test:')
df_test = adfuller(data, autolag='AIC')
df_output = pd.Series(df_test[0:4],
                    index=['Test Statistic','p-value','#Lags Used','Number of Observations Used']
                    )
for key,value in df_test[4].items():
    df_output['Critical Value (%s)'%key] = value
print(df_output)
```
